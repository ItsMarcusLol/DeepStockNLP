```python

all_head = []
for y in years:
    for m in months:
        if m == 12:
            d2 = 31
            m2 = m
        else:
            d2 = 1
            m2 = m + 1
        url = 'https://globalnews.xignite.com/xGlobalNews.json/GetHistoricalSecurityHeadlines?IdentifierType=Symbol&Identifier='+stock_ticker+'&StartDate='+str(m)+'/'+str(d1)+'/'+str(y)+'&EndDate='+str(m2)+'/'+str(d2)+'/'+str(y)+'&_token='+your_key
        r = requests.get(url)
        json_data = r.json()
        data = json_data['Headlines']
        for x in data:
            h = x['Title'].lower()
            if stock_name in h or company_name in h:
                if h not in all_head:
                    oldformat = x['Date']
                    d = datetime.strptime(oldformat, '%m/%d/%Y').date()
                    all_head.append((d,h))
```

```python
df = pd.DataFrame(all_head, columns = ["Date", "Headline"])
print(df)
```
